## ResNet architectures

__Question 3 (2 points) :__ Write a classification pipeline for $\mathcal{X}_{\text{train}}$, train from scratch and evaluate a *ResNet-18* architecture specific to CIFAR10 (details about the ImageNet model can be found here: https://arxiv.org/abs/1512.03385). Please report the accuracy obtained on the whole dataset as well as the reference paper/GitHub link.

*Hint :* You can re-use the following code : https://github.com/kuangliu/pytorch-cifar. During a training of 10 epochs, a batch size of 10 and a learning rate of 0.01, one obtains 40% accuracy on $\mathcal{X}_{\text{train}}$ (\~2 minutes) and 20% accuracy on $\mathcal{X}_{\text{test}}$ (\~5 minutes).

On reprend le code de kuangliu pour le Resnet 18 puis on construit le pipeline de train et test. Après plusieurs essais on remarque que l'optimizer SGD présente de meilleures performances que Adam dans notre cas.

```python
###################
# kuangliu RESNET #
###################

class BasicBlock(nn.Module):
    expansion = 1

    def __init__(self, in_planes, planes, stride=1):
        super(BasicBlock, self).__init__()
        self.conv1 = nn.Conv2d(
            in_planes, planes, kernel_size=3, stride=stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                               stride=1, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)

        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != self.expansion*planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, self.expansion*planes,
                          kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion*planes)
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out


class Bottleneck(nn.Module):
    expansion = 4

    def __init__(self, in_planes, planes, stride=1):
        super(Bottleneck, self).__init__()
        self.conv1 = nn.Conv2d(in_planes, planes, kernel_size=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                               stride=stride, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)
        self.conv3 = nn.Conv2d(planes, self.expansion *
                               planes, kernel_size=1, bias=False)
        self.bn3 = nn.BatchNorm2d(self.expansion*planes)

        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != self.expansion*planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, self.expansion*planes,
                          kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion*planes)
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = F.relu(self.bn2(self.conv2(out)))
        out = self.bn3(self.conv3(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out


class ResNet(nn.Module):
    def __init__(self, block, num_blocks, num_classes=10):
        super(ResNet, self).__init__()
        self.in_planes = 64

        self.conv1 = nn.Conv2d(3, 64, kernel_size=3,
                               stride=1, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(64)
        self.layer1 = self._make_layer(block, 64, num_blocks[0], stride=1)
        self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
        self.layer3 = self._make_layer(block, 256, num_blocks[2], stride=2)
        self.layer4 = self._make_layer(block, 512, num_blocks[3], stride=2)
        self.linear = nn.Linear(512*block.expansion, num_classes)

    def _make_layer(self, block, planes, num_blocks, stride):
        strides = [stride] + [1]*(num_blocks-1)
        layers = []
        for stride in strides:
            layers.append(block(self.in_planes, planes, stride))
            self.in_planes = planes * block.expansion
        return nn.Sequential(*layers)

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = F.avg_pool2d(out, 4)
        out = out.view(out.size(0), -1)
        out = self.linear(out)
        return out


def ResNet18():
    return ResNet(BasicBlock, [2, 2, 2, 2])

##################
# Train pipeline #
##################

def train(
    model, 
    epochs, 
    optimizer,
    scheduler,
    f_score, 
    train_loader, 
    verbose=True, 
    device=device, 
    start_epoch=0
):

  model = model.to(device)
  # On garde en mémoire la moyenne de la loss sur des batch et l'accuracy
  train_loss = []
  train_accuracy = []
  for epoch in range(start_epoch, epochs+start_epoch): 
    # On initialise le model pour le train et les métriques associées au batch
    model.train()

    # La loss moyenne du batch
    sum_loss = 0
    
    # Pour l'accuracy
    n_samples = 0
    n_correct = 0   
    
    for batch_idx, (data, target) in enumerate(train_loader):
        
        # on récupère le batch et on le met sur le bon device
        data, target = data.to(device), target.to(device) 
        # on initialise les gradients avant le forward
        optimizer.zero_grad()
        output = model(data)
        # on calcule la loss puis on fait le backward
        loss = f_score(output, target)
        loss.backward()
        # on met à jours les paramètres et le sceduler
        optimizer.step()
        scheduler.step()
        # on garde en mémoire la loss pour le calcul de la loss moyenne
        sum_loss += loss.cpu().item()
        
        ## Pour l'accuracy du batch :
        # on choisit la classe la plus probable
        y_pred = output.argmax(dim=1)
        # On incrémente le nombre de sample et la 0-1 loss pour la calcul de l'accuracy:
        n_samples += len(target)
        n_correct += (y_pred == target).sum().cpu().item()

    # On calcule l'accuracy du batch et sa loss moyenne
    accuracy = np.round(n_correct /n_samples, 3)
    mean_loss = np.round(sum_loss/n_samples, 3)
    # On garde en mémoire ces métriques
    train_accuracy.append(accuracy)
    train_loss.append(mean_loss)

    if verbose :
      # on print l'accuracy au fûr et à mesure de l'entraînement
      if epoch % 3 == 0  or epoch == epochs+start_epoch - 1 : 
        print(f"Train Epoch: {epoch} | Mean Loss={mean_loss}, Accuracy={accuracy}")
  return train_accuracy , train_loss

def plot_train(train_accuracy, train_loss):
  fig, ax = plt.subplots(1,2 , figsize=(10,3))
  ax[0].plot(np.arange(len(train_accuracy)), train_accuracy)
  ax[0].set_title('Train Accuracy')
  ax[0].set_xlabel('epochs')
  ax[1].plot(np.arange(len(train_loss)), train_loss)
  ax[1].set_title('Train Loss')
  ax[1].set_xlabel('epochs')
  plt.show()


#################
# Test pipeline #
#################

def test (
    model, 
    f_score, 
    test_loader,
    verbose=True,
    device=device
):
  model.eval()
  # On calcul la moyenne des loss et l'accuracy par batch
  with torch.no_grad():
    # La loss moyenne du batch
    sum_loss = 0 
    # Pour l'accuracy
    n_samples = 0
    n_correct = 0  
    # On récupère le batch et on le met sur le bon device
    for batch_idx, (data, target) in enumerate(test_loader):
      data, target = data.to(device), target.to(device)
      # Calcul de la loss
      output = model(data)
      loss = f_score(output, target)
      # on garde en mémoire la loss pour le calcul de la loss moyenne
      sum_loss += loss.cpu().item()
        
      ## Pour l'accuracy :
      # on choisit la classe la plus probable
      y_pred = output.argmax(dim=1)
      # On incrémente le nombre de sample et la 0-1 loss:
      n_samples += len(target)
      n_correct += (y_pred == target).sum().cpu().item()

  # On calcule l'accuracy et la loss moyenne
  accuracy = np.round(n_correct/n_samples, 3)
  mean_loss = np.round(sum_loss/n_samples, 3)
  if verbose :
      print(f"TEST | Mean Loss={mean_loss}, Accuracy={accuracy}")
  return  accuracy


#####################
# standard pipeline #
#####################

def standard_train_test(
    train_loader,
    model,
    optimizer,
    device, 
    epochs = 10,
    ):
    scheduler = torch.optim.lr_scheduler.StepLR(
            optimizer, 
            step_size=30, # décroit tous les 30
            gamma=0.1 # divise par 10 le lr
        )
    f_score = torch.nn.CrossEntropyLoss()
    ## Train
    train_accuracy , train_loss = train(
        model, 
        epochs, 
        optimizer,
        scheduler,
        f_score, 
        train_loader, 
        verbose=True, 
        device=device, 
        start_epoch=0
    )
    plot_train(train_accuracy, train_loss)
    ## Test
    test_loader = torch.utils.data.DataLoader(cifar10_test, batch_size=10)
    test_accuracy = test(
        model, 
        f_score, 
        test_loader,
        verbose=True,
        device=device
    )

##########
# Script #
##########

## Définition des loaders
train_loader_baseline = torch.utils.data.DataLoader(cifar10_train_100, batch_size=10, shuffle=True)
model_baseline = ResNet18()
optimizer_baseline = torch.optim.SGD(model_baseline.parameters(), lr=1e-2, momentum=0.8)
# optimizer_baseline = torch.optim.Adam(model_baseline.parameters(), lr=1e-2)
standard_train_test(
    train_loader_baseline,
    model_baseline,
    optimizer_baseline,
    device, 
    epochs = 10
    )
```

Comme on s'en doutait, on observe un problème d'overfitting. On peut alors comparer avec un entraînement sur la base complète :

```python
###################
# Script all data #
###################
## Définition des loaders
train_loader_all = torch.utils.data.DataLoader(cifar10_train, batch_size=10, shuffle=True)
model_all = ResNet18()
optimizer_all = torch.optim.SGD(model_all.parameters(), lr=1e-2, momentum=0.8)
standard_train_test(
    train_loader_all,
    model_all,
    optimizer_all,
    device, 
    epochs = 10, 
    )
```

__Question 4 (1 points):__ Pick a model from the list above, adapt it for CIFAR10 and retrain its final layer (or a block of layers, depending on the resources to which you have access to). Report its accuracy.

On adapte un ResNet préentrainé en ajoutant des couches denses afin de classifier sur 10 classes. Plusieurs architectures et stratégies ont été testés:
- Réentrainer tous les paramètres.
- Réentrainer seulement les paramètres de la couche dense.
- Réentrainer les paramètres de la couche dense puis tous les paramètres.

```python
#####################
# pretrained Resnet #
#####################

from torchvision.models import resnet18, resnet34, resnet50, resnet101

class pt_ResNet(nn.Module):
    def __init__(self, net = resnet50):
        super().__init__()
        self.model = net(pretrained=True)
        self.num_filters = self.model.fc.in_features
        self.layers = list(self.model.children())[:-1]
        self.backbone = nn.Sequential(*self.layers)
        # On ajoute des couches denses pour réaliser la classification à 10 classes
        self.fc1 = nn.Linear(self.num_filters, self.num_filters//2)
        self.fc2 = nn.Linear(self.num_filters//2, 10)

    def forward(self, x):
      x = self.backbone(x)
      x = x.squeeze(3)
      x = x.squeeze(2)
      x = self.fc1(x)
      x = self.fc2(x)
      return x
        
    #on ajoute une méthode pour éviter le réapprentissage sur la partie déjà entrainé
    def freeze_backbone(self):
        for param in self.model.parameters():
            param.requires_grad = False
            
        for param in self.fc1.parameters():
            param.requires_grad = True
            
        for param in self.fc2.parameters():
            param.requires_grad = True
    
    def unfreeze_backbone(self):
        for param in self.model.parameters():
            param.requires_grad = True

##########
# Script #
##########

nepochs = 10
lst_model = [pt_ResNet(resnet18), pt_ResNet(resnet34), pt_ResNet(resnet50), pt_ResNet(resnet101)]
lst_model2 = [pt_ResNet(resnet18), pt_ResNet(resnet34), pt_ResNet(resnet50), pt_ResNet(resnet101)]

lst_names = [f"ResNet{k}" for k in [18,34,50,101]]

for i, model_pretrained in enumerate(lst_model):
    name = lst_names[i]
    print(f"\n_____{name} all parameters____")
    optimizer_pretrained = torch.optim.SGD(model_pretrained.parameters(), lr=1e-2, momentum=0.8)
    
    standard_train_test(
        torch.utils.data.DataLoader(cifar10_train_100, batch_size=10, shuffle=True),
        model_pretrained,
        optimizer_pretrained,
        device, 
        epochs = nepochs, 
    )

    print(f"\n_____{name} fc parameters____")
    model_pretrained2 = lst_model2[i]
    model_pretrained2.freeze_backbone()
    optimizer_pretrained2 = torch.optim.SGD(model_pretrained2.parameters(), lr=1e-2, momentum=0.8)
    
    standard_train_test(
        torch.utils.data.DataLoader(cifar10_train_100, batch_size=10, shuffle=True),
        model_pretrained2,
        optimizer_pretrained2,
        device, 
        epochs = nepochs, 
    )
    print(f"\n_____{name} fc parameters + train all____")
    model_pretrained2.unfreeze_backbone()
    optimizer_pretrained3 = torch.optim.SGD(model_pretrained2.parameters(), lr=1e-2, momentum=0.8)
    standard_train_test(
        torch.utils.data.DataLoader(cifar10_train_100, batch_size=10, shuffle=True),
        model_pretrained2,
        optimizer_pretrained3,
        device, 
        epochs = nepochs, 
    )
```

# Incorporating *a priori*
Geometrical *a priori* are appealing for image classification tasks, though one might have to handle several boundary effects.

__Question 5 (0.5 points) :__ Explain the issues when dealing with translations, rotations, scaling effects, color changes on $32\times32$ images. Propose several ideas to tackle them.

Les transformations des données pour la 'data augmentation' conduisent en générale à une perte d'information d'une partie de l'image. Pour la translation les pixels déplacés en dehors des limites de l'image sont ignorés, et dans la rotation les pixels autour des coins peuvent ne pas être alignées avec la grille de l'image originale, créant des vides. Aussi, le changement d'échelle floutte l'image (perte de résolution) et les modification de couleurs peuvent modifier les contrastes et l'intensité lumineuse. 

Pour y remédier, il existe différentes astuces possibles: 
- On interpole les pixels manquants, ou on se contente de symmétries et de "crop".
- On applique un algorithme de surésolution pour se ramener à des spectres comparables entre nos images initiales et modifiées. Un antialiasing peut par exemple être utilisé.
- On réalise la modification de couleur dans l'espace HSV afin de conserver l'intensité lumineuse V.

## Data augmentations

__Question 6 (4 points):__ Propose a set of geometric transformation beyond translation, and incorporate them in your training pipeline. Train the model of the __Question 3__ with them and report the accuracies.

Dans cette partie on teste les transformations standards suivantes:
- RandomHorizontalFlip: symmétrie par rapport à l'axe vertical.
- RandomCrop: Zoom dans l'image (traduit une translation et un padding).
- GaussianBlur: Floute l'image de manière à conserver les basses fréquences.
- ColorJitter: Modifie la coordonné H dans la représentation HSV (changement de teinte).
- AutoAugmentPolicy.CIFAR10: transformation des données tirée de https://arxiv.org/pdf/1805.09501.pdf .

```python
##############################
# pipeline data augmentation #
##############################

def get_train_loader(lst_transform, add_transform = True):
    if lst_transform != []:
        transformed_cifar10_train = torchvision.datasets.CIFAR10(
            root ="\content", 
            train=True, # train set
            download=True , # copie en locale
            transform=v2.Compose(lst_transform+[v2.ToTensor()])
        )
        tranformed_cifar10_train_100 = torch.utils.data.Subset(transformed_cifar10_train, list(np.arange(100)))
        
        if add_transform:
            increased_dataset = torch.utils.data.ConcatDataset([tranformed_cifar10_train_100, cifar10_train_100])
        else:
            increased_dataset = tranformed_cifar10_train_100
        
        return torch.utils.data.DataLoader(increased_dataset, batch_size=10, shuffle=True)
        
    return torch.utils.data.DataLoader(cifar10_train_100, batch_size=10, shuffle=True)

def data_augment_train_test_pipeline(
    lst_transform=[], 
    train_id = "no augmentation", 
    device=device, 
    add_transform =True
):
    nepochs = 10
    dct_model = {
        "pt_ResNet18":pt_ResNet(resnet18),
        "pt_ResNet34":pt_ResNet(resnet34)
    }
    for name, model in dct_model.items():
        print(f"\n______{name}|{train_id}_____")
        optimizer = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.8)
        standard_train_test(
            get_train_loader(lst_transform, add_transform=add_transform),
            model,
            optimizer,
            device, 
            epochs = nepochs, 
        )

##########
# Script #
##########
dct_transform={
    "no augmentation": [],
    "flip":[
        v2.RandomHorizontalFlip(p=1), 

    ],
    "flip,crop":[
        v2.RandomHorizontalFlip(p=1), 
        v2.RandomCrop(size=[32,32], padding=3),

    ],
    "flip,blur":[
        v2.RandomHorizontalFlip(p=1), 
        v2.GaussianBlur(kernel_size=(3,3)),

    ],
    "flip,crop,blur":[
        v2.RandomHorizontalFlip(p=1), 
        v2.RandomCrop(size=[32,32], padding=3),
        v2.GaussianBlur(kernel_size=(3,3)),
    ],
    "flip,crop,blur,color":[
        v2.RandomHorizontalFlip(p=1), 
        v2.RandomCrop(size=[32,32], padding=3),
        v2.GaussianBlur(kernel_size=(3,3)),
        v2.ColorJitter(hue=.1)
    ],
    "Auto Augmentation": [v2.AutoAugment(v2.AutoAugmentPolicy.CIFAR10)],
}

for train_id, lst_transform in dct_transform.items():
    data_augment_train_test_pipeline(lst_transform, train_id, device)
```

# Weak supervision

__Bonus \[open\] question (up to 3 points) :__ Pick a weakly supervised method that will potentially use $\mathcal{X}\cup\mathcal{X}_{\text{train}}$ to train a representation (a subset of $\mathcal{X}$ is also fine). Evaluate it and report the accuracies. You should be careful in the choice of your method, in order to avoid heavy computational effort.

Trouver une version différentiable d'un "manifold regularization" n'est pas évident, et afin de comparer l'approche suivante à l'étude précédente nous n'allons pas modifier l'architecture du modèle. Ainsi, nous n'allons pas étudier les méthodes de **Contrastive Learning** comme dans https://arxiv.org/pdf/2002.05709.pdf, mais se focaliser sur une approche plus simple : le **retraining** ou **pseudo labelling**.
Cette approche est connu depuis longtemps et présente de nombreuse limitations: dans https://arxiv.org/pdf/1908.02983.pdf les auteurs relèvent un risque d'overfitting en raison d'un 'biais de confirmation' inhérent à la méthode.
Certaines évolutions de cette méthodes ont permis de grandement améliorer les performances sur Cifar10 comme le **Meta Pseudo Labels** (https://arxiv.org/pdf/2003.10580v4.pdf), hélas ce dernier reposant sur le paradygme "Teacher Student" nécessiterait plus de temps pour être étudiée. 
Finalement, nous allons étudier le **Curriculum Labeling** (https://arxiv.org/pdf/2001.06001.pdf) qui filtre simplement les nouvelles labellisation en fonction du score de labellisation. Ainsi, si le modèle "hésite" entre deux classes, la prédiction ne sera pas utilisée pour l'entrainement du cycle d'après. 


```python
#############################
# pipeline pseudo labelling #
#############################

batch_size = 10

def unalabeled_idx(idx, ntrain):
    return list(set(np.arange(ntrain)) - set(idx))

def get_unlabeled_loader(idx, train_dataset):
    ntrain = len(train_dataset)
    un_idx = unalabeled_idx(idx, ntrain)
    train_dataset_unlabeled = torch.utils.data.Subset(train_dataset, un_idx)
    return torch.utils.data.DataLoader(train_dataset_unlabeled, batch_size=batch_size, shuffle=True), un_idx

def get_l_and_un_loaders(idx, train_dataset):
    train_labeled_dataset = torch.utils.data.Subset(train_dataset, idx)
    train_l_loader = torch.utils.data.DataLoader(train_labeled_dataset, batch_size=batch_size, shuffle=True)
    train_un_loader, un_idx = get_unlabeled_loader(idx, train_dataset)
    return train_l_loader, train_un_loader, un_idx

def max_proba(batch_outputs):
    return torch.max(F.softmax(batch_outputs,dim=1),dim=1).values.cpu().numpy()

def test_unlabeled (
    model, 
    f_score, 
    unlabeled_loader,
    un_idx,
    threshold = 0.8, 
    device=device
):
  model.to(device)
  new_idx = []
  model.eval()
  # On calcul la moyenne des loss et l'accuracy par batch
  with torch.no_grad():
    # La loss moyenne du batch
    sum_loss = 0 
    # Pour l'accuracy
    n_samples = 0
    n_correct = 0  
    # On récupère le batch et on le met sur le bon device
    for i, (data, target) in enumerate(unlabeled_loader):
      indx = un_idx[i*batch_size:(i+1)*batch_size]
      data, target = data.to(device), target.to(device)
      # Calcul de la loss
      output = model(data)
      mproba = max_proba(output)
      new_idx += list(np.array(indx)[mproba>threshold])
        
      loss = f_score(output, target)
      # on garde en mémoire la loss pour le calcul de la loss moyenne
      sum_loss += loss.cpu().item()
      
      ## Pour l'accuracy :
      # on choisit la classe la plus probable
      y_pred = output.argmax(dim=1)
      # On incrémente le nombre de sample et la 0-1 loss:
      n_samples += len(target)
      n_correct += (y_pred == target).sum().cpu().item()

  # On calcule l'accuracy et la loss moyenne
  accuracy = np.round(n_correct/n_samples, 3)
  mean_loss = np.round(sum_loss/n_samples, 3)
  print(f"Unlabeled TEST | Mean Loss={mean_loss}, Accuracy={accuracy}")
  return  new_idx

def pseudo_labelling_train_test(
    train_dataset,
    model,
    device,
    labeled_idx = list(np.arange(100)),
    epochs = 10,
    nlabelling = 1
    ):
    idx = labeled_idx.copy()
    thrs = 0.8
    
    for k in range(nlabelling+1):
        
        # Data selection
        train_l_loader, train_un_loader, un_idx = get_l_and_un_loaders(idx, train_dataset)
        # Train init
        optimizer = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.8)
        scheduler = torch.optim.lr_scheduler.StepLR(
                optimizer, 
                step_size=30, # décroit tous les 30
                gamma=0.1 # divise par 10 le lr
            )
        f_score = torch.nn.CrossEntropyLoss()
        
        ## Train
        print(f"Training {k}")
        train_accuracy , train_loss = train(
            model, 
            epochs, 
            optimizer,
            scheduler,
            f_score, 
            train_l_loader, 
            verbose=True, 
            device=device, 
            start_epoch=0
        )
        if k < nlabelling:
            ## Test on unlabeled
            new_idx = test_unlabeled (
                model, 
                f_score, 
                train_un_loader,
                un_idx,
                threshold = thrs, 
                device=device
            )
            print(f"number of new pseudo labels {len(new_idx)}")
            
            idx += new_idx
            
            thrs -=0.1
    ## Test
    test_loader = torch.utils.data.DataLoader(cifar10_test, batch_size=10)
    test_accuracy = test(
        model, 
        f_score, 
        test_loader,
        verbose=True,
        device=device
    )
```
